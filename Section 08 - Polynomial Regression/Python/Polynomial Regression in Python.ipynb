# Polynomial Regression in Python
<hr>

## Importing the Libraries

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

<hr>

## Importing the Dataset

```python
dataset = pd.read_csv("Position_Salaries.csv")
X = dataset.iloc[:, 1:-1].values
y = dataset.iloc[:, -1].values
```

<hr>

## Training the Linear Regression model on the whole dataset

* *Linear Regression* is a statistical method used to model the relationship between a dependent variable and one or more independent variables. It is a type of supervised learning algorithm that is used to predict a continous outcome variable based on one or more predictor variables.

* The goal of Linear Regression is to find the best-fitting line or hyperplane that represents the relationship between the independent and dependent variables. This line can be used to make predictions about future outcomes based on the values of the independent variables.

* Linear Regression is widely used in various fields such as finance, economics, and machine learning. It is used to analyze and predict the relationship between various variables such as sales, marketing, and customer behavior. It is also used to identify the most important factors that influence a dependent variable and can be used to reduce the complexity of a dataset.

* One of the main advantages of Linear Regression is that it is easy to understand and interpret. It provides a clear and concise explanation of the relationship between the dependent and independent variables, making it easy for stakeholders to understand and make informed decisions.

* Another advantage of Linear Regression is that it is relatively simple to implement and requires minimal preprocessing of the data. It can be used with a variety of programming languages and tools, making it accessible to a wide range of users.

* In conclusion, Linear Regression is a powerful and widely used statistical method that is used to model the relationship between a dependent variable and one or more independent variables. It is easy to understand and interpret, and can be implemented with minimal preprocessing of the data.

```python
from sklearn.linear_model import LinearRegression

lin_reg = LinearRegression()
lin_reg.fit(X, y)
```

<hr>

## Training the Polynomial Regression model on the whole dataset

```python
from sklearn.preprocessing import PolynomialFeatures

# poly_reg = PolynomialFeatures(degree = 2)
poly_reg = PolynomialFeatures(degree = 4)
X_poly = poly_reg.fit_transform(X)
lin_reg_2 = LinearRegression()
lin_reg_2.fit(X_poly, y)
```

<hr>

## Visualizing the Linear Regression results

```python
plt.scatter(X, y, color = "red")
plt.plot(X, lin_reg.predict(X), color ="blue")
plt.title("Truth or Bluff (Linear Regression)")
plt.xlabel("Postion Level")
plt.ylabel("Salary")
plt.show()
```

<hr>

## Visualizing the Polynomial Regression results

```python
plt.scatter(X, y, color = "red")
plt.plot(X, lin_reg_2.predict(X_poly), color ="blue")
plt.title("Truth or Bluff (Polynomial Regression)")
plt.xlabel("Postion Level")
plt.ylabel("Salary")
plt.show()
```

<hr>

## Visualizing the Polynomial Regression results (for higher resolution and smoother curve)

```python
X_grid = np.arange(min(X), max(X), 0.1)
X_grid = X_grid.reshape((len(X_grid), 1))
plt.scatter(X, y, color = "red")
plt.plot(X_grid, lin_reg_2.predict(poly_reg.fit_transform(X_grid)), color = "blue")
plt.title("Truths or Bluff (Polynomial Regression)")
plt.xlabel("Position Level")
plt.ylabel("Salary")
plt.show()
```

<hr>

## Predicting a new result with Linear Regression

```python
lin_reg.predict([[6.5]])
```

<hr>

## Predicting a new result with Polynomial Regression

```python
lin_reg_2.predict(poly_reg.fit_transform([[6.5]]))
```
